```python
import time
global preguntas
global datos_preguntas
global respuestas_correctas
global respuestas_incorrectas
global lista_notas


preguntas = [['1', 'Matemáticas', '1+1', '1', '2', '3', '4', '2' ],
             ['2', 'Matemáticas', '2+2', '1', '2', '3', '4', '4' ],
             ['3', 'Matemáticas', '2-2', '0', '1', '3', '4', '0' ],
             ['4', 'Matemáticas', '2/2', '0', '1', '3', '4', '1' ],
             ['5', 'Lectura', 'Sinónimo de rápido', 'Lento', 'Velóz', 'Largo', 'Corto', 'Velóz' ],
             ['6', 'Lectura', '¿Cuántas sílabas tiene la palabra TACO', '1', '2', '3', 'Ninguna', '2' ],
             ['7', 'Lectura', '¿Cuántas sílabas tiene la palabra COMPUTADORA', '2', '3', '5', '7', '5' ],
             ['8', 'Ciencias', 'Proceso de respiración de las plantas', 'Reproducción sexual', 'Reproducción asexual', 'Fotosíntesis', 'Respiración', 'Fotosíntesis' ],
             ['9', 'Ciencias', 'Metodología utilizada en la investigación', 'Ensayo', 'Reporte', 'Cuento', 'Metodo científico', 'Metodo científico' ],
             ['10', 'Ciencias', 'Planeta más cercano al Sol', 'Tierra', 'Venus', 'Mercurio', 'Marte', 'Mercurio' ]]


datos_preguntas = ["Ingresa el ID de la pregunta: ",
                     "Ingresa el número del área de la pregunta : 1.- Lectura, 2.- Matemáticas, 3.- Ciencias: ",
                     "Ingresa el texto de la pregunta: ",
                     "Ingresa la respuesta 1: ",
                     "Ingresa la respuesta 2: ",
                     "Ingresa la respuesta 3: ",
                     "Ingresa la respuesta 4: ",
                     "Ingresa la respuesta correcta: ",]

respuestas_correctas = 0
respuestas_incorrectas = 0
lista_notas = []

def registrar_pregunta():
  
  pregunta = []
  dato = 0
  while dato < len(datos_preguntas):
    
    dato_nuevo = input(datos_preguntas[dato])
    
    if dato == 1:
      if int(dato_nuevo) >= 1 and int(dato_nuevo) <= 3:
        pregunta.append(int(dato_nuevo))
      else: 
        print("Ingresa un número correcto")
        dato -= 1
    else:
      pregunta.append(dato_nuevo)
    dato += 1

  preguntas.append(pregunta)


def actualizar_pregunta():

  id_cambio = input("Ingresa el ID de la pregunta a actualizar: ")
 
  for num in range(len(preguntas)):
    
    pregunta = preguntas[num]

    id_pregunta = pregunta[0]
    
    if id_cambio == id_pregunta:
      existe = True
      no_pregunta = num
      break
    else:
      existe = False
  
  if existe == True:
    print("Si existe la pregunta")
      
    id, area, texto_pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = preguntas[no_pregunta]


    if area == 1:
      area = "Lectura"
    elif area == 2:
      area = "Matemáticas"
    else:
      area = "Ciencias"
  
    print(f'{"ID: "}{id:<5}{"Área: "}{area}\n{texto_pregunta}\n{respuesta1:<15}{respuesta2:<15}{respuesta3:<15}{respuesta4}\n{"Respuesta correcta: "}{correcta}')
   
    confirmacion = input("Si deseas cambiar la pregunta escribe 'si' ")

    if confirmacion.lower() == "si":
     
      pregunta = []
      dato = 0
      while dato < len(datos_preguntas):
        
        dato_nuevo = input(datos_preguntas[dato])
        
        if dato == 1:
          if int(dato_nuevo) >= 1 and int(dato_nuevo) <= 3:
            pregunta.append(int(dato_nuevo))
          else: 
            print("Ingresa un número correcto")
            dato -= 1
        else:
          pregunta.append(dato_nuevo)
        dato += 1
      
      id, area, texto_pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta
      
      print(f'{"Los datos de la pregunta a actualizar son:"}\n{"ID: "}{id:<5}{"Área: "}{area}\n{texto_pregunta}\n{respuesta1:<15}{respuesta2:<15}{respuesta3:<15}{respuesta4}\n{"Respuesta correcta: "}{correcta}')
      
      confirmacion = input("Si deseas cambiar la pregunta escribe 'si' ")

      if confirmacion.lower() == "si":
        preguntas[no_pregunta] = pregunta
    else:
      print("No existe la pregunta, intenta con otro ID")
      actualizar_pregunta()

def estudiar_area():
  global preguntas
  print("Bienvenido a la sección de estudio")
  area = int(input("Ingresa el área que deseas estudiar  1.- Lectura, 2.- Matemáticas, 3.- Ciencias: "))
  
  if area >= 1 and area <= 3:
   
    lista_area = []
   
    for pregunta in preguntas:
      if pregunta[1] == area:
        lista_area.append(pregunta)

    num_preguntas = int(input(f'{"Ingresa el número de preguntas que deseas estudiar, debe mayor a cero y menor o igual que "}{len(lista_area)}: '))

    if num_preguntas < 0 or num_preguntas > len(lista_area):
      print("Ingresa un número correcto de preguntas: ")
      estudiar_area()
    else:
      for num in range(num_preguntas):
        
        pregunta = lista_area[num]

        id, area, texto_pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta
        
        if area == 1:
          area = "Lectura"
        elif area == 2:
          area = "Matemáticas"
        else:
          area = "Ciencias"
    
        print(f'{"Área: "}{area}\n{texto_pregunta}\n{respuesta1:<15}{respuesta2:<15}{respuesta3:<15}{respuesta4}\n{"Respuesta correcta: "}{correcta}\n')
        
  else: 
    print("Ingresa un número correcto")
    estudiar_area()

def presentar_quiz():
    global preguntas
    global respuestas_correctas
    global respuestas_incorrectas
    global lista_notas

    print("Bienvenido al Quiz \nPor favor escribe la respuesta correcta en el cuadro de texto cuándo se indique")

    for num in range(10):
      tiempo = 1200
      inicio = time.time()
      
      pregunta = preguntas[num]

      id, area, texto_pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta
      
      if area == 1:
        area = "Lectura"
      elif area == 2:
        area = "Matemáticas"
      else:
        area = "Ciencias"
  
      print(f'{texto_pregunta}\n{respuesta1:<15}{respuesta2:<15}{respuesta3:<15}{respuesta4}\n')
      respuesta = input("Escribe la respuesta correcta: ")
      if respuesta == correcta:
        respuestas_correctas += 1
      else:
        respuestas_incorrectas += 1
      
      fin = time.time()

      total = round((tiempo - (fin - inicio)) / 60, 2)
      print(f"\nEl tiempo restante son {total} minutos\n")
    
    calificacion = respuestas_correctas * 10

    print(f"Tu calificación es: {calificacion}")
    lista_notas.append(calificacion)

def reportar_calificaciones():
  global respuestas_correctas
  global respuestas_incorrectas
  global lista_notas

  usuarios = len(lista_notas)
  
  suma = 0
  for nota in lista_notas:
    suma = suma + nota
  
  promedio = suma / len(lista_notas)

  porcentaje_correctas = round(respuestas_correctas / (respuestas_correctas + respuestas_incorrectas) * 100, 2)
  porcentaje_incorrectas = round(respuestas_incorrectas / (respuestas_correctas + respuestas_incorrectas) * 100, 2)
  
  print(f'{"Total de usuarios":<30}{usuarios}\n{"Promedio de calificaciones":<30}{promedio}\n{"Respuestas correctas":<30}{porcentaje_correctas}%\n{"Respuestas incorrectas":<30}{porcentaje_incorrectas}%')



def menu(): 
  print("A- Registrar pregunta")
  print("B- Actualizar pregunta")
  print("C- Estudiar área")
  print("D- Presentar quiz")
  print("E- Reportar calificaciones")
  print("F- Salir")
  
  opcion = input("Seleccione una opción:")
  
  return opcion

def salir():
  print("Gracias por usar el sistema")
  sigue = False

  return sigue

def main():

  sigue = True
  while sigue:

    opcion = menu()

    if opcion not in ["A","B","C","D","E","F"]:
      print("Selecciona una letra del menú")
      continue 
    
    elif opcion == "A": 
      registrar_pregunta()

    elif opcion == "B":
      print("Actualizar pregunta")
      actualizar_pregunta()

    elif opcion == "C":
      print("Estudiar área")
      estudiar_area()

    elif opcion == "D":
      print("Presentar quiz")
      presentar_quiz()

    elif opcion == "E":
      print("Reportar calificaciones")
      reportar_calificaciones()

    else:
      sigue = salir()
        
main()
```
